```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline
```

# Read the data and Explore the Data

```python
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

# Removing Missing Values

Matplotlib expects the columns of values we pass in have matching lengths and missing value will cause matplotlib to throw errors.

Only one row contained missing values and was dropped

```python
raw_data_count = len(recent_grads)

recent_grads=recent_grads.dropna()
cleaned_data_count = len(recent_grads)

print("Number of rows before: ",raw_data_count,"\n", "Number of rows after: ",cleaned_data_count)
```

# Examining Relationships Using Scatter Plots

Plots to evaluate the relationships between several columns will be used to answer the following questions

1. Do students in more popular majors make more money?
2. Do students that majored in subjects that were majority female make more money?
3. Is there any link between the number of full-time employees and median salary?

### Sample_size and Median

```python
ax = recent_grads.plot(x='Sample_size',y="Median",title = "Sample Size vs Median Pay",kind='scatter')
```

### Sample_size and Unemployment_rate

```python
ax1 = recent_grads.plot(x='Sample_size',y="Unemployment_rate",title = "Sample Size vs Unemployment Rate",kind='scatter')
```

### Full_time and Median

```python
ax2 = recent_grads.plot(x='Full_time',y="Median",title = " # of Full Time Jobs vs Median Pay",kind='scatter')
```

### ShareWomen and Unemployment Rate

```python
ax3 = recent_grads.plot(x='ShareWomen',y="Unemployment_rate",title = "% of Women in Major vs Unemployment Rate",kind='scatter')
```

### ShareWomen and Median

```python
ax4 = recent_grads.plot(x='ShareWomen',y="Median",title = "% of Women in Major vs Median Pay",kind='scatter')
```

### Men and Median

```python
ax5 = recent_grads.plot(x='Men',y="Median",title = "# of Men in Major vs Median Pay",kind='scatter')
```

### Women and Median

```python
ax6 = recent_grads.plot(x='Women',y="Median",title = "# of Women in Major vs Median Pay",kind='scatter')
```

## Observations

1. Students in Popular Majors do not make more money. Based on the first plot showing sample size vs median pay, the majors that had larger sample sizes had lower pay than those with higher median salaries. However, majors with large sample sizes also had lower unemployment rates than those majors with smaller sample sizes. A trade-off between picking a popular major could be job security vs. pay.
2.  There is a slight negative correlation in ShareWomen vs Median pay. Students who majored in subjects that were majority female do not make more money.
3. There is no clear correlation between the number of full-time eployees and median salaries.

3. Is there any link between the number of full-time employees and median salary?
